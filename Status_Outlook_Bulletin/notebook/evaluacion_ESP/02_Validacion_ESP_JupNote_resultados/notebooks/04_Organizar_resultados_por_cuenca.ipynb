```python
import pandas as pd
import os
```

```python
ruta = r"../Resultados/02_Tabla_de_metricas_por_mes"

dfs = []
```

```python
for archivo in os.listdir(ruta):
    if archivo.endswith(".csv"):
        path = os.path.join(ruta, archivo)
        
        df = pd.read_csv(path)

        df = df[df["lead"].isin([1])]
        
        df = df.groupby("mes").mean()

        df = df.drop(columns=["NS","Brier_Q90","CV","Pearson","CRPSSad","Brier_Q10","Brier_Q25_Q75","Brier_Q75_Q90","Brier_Q10_Q25","N_det","N_prob"])

        cuenca = os.path.basename(archivo).split("_")[0]
        
        df = df.rename(columns={"Pbias": cuenca})
        
        dfs.append(df)

# Concatenar por columnas (alineando por lead)

df_final = pd.concat(dfs, axis=1)
#print (df_final)
df_final = df_final.transpose()

# Guardar
df_final.to_csv(os.path.join(r"../Resultados/Metricas_por_cuenca/datos_Pbias_por_cuenca_por_mes_lead1.csv"))
```
