```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
ruta_carpeta = Path(
    "../Resultados/Tabla_de_metricas_por_mes_por_cuenca"
)

ruta_salida = Path(
    "../Resultados/Graficos_por_mes"
)
ruta_salida.mkdir(parents=True, exist_ok=True)
```

```python
df = pd.read_csv(f"{ruta_carpeta}/{cuenca}_df_esp_con_historico_crpss_bootstrap.csv")

lead_times_evaluar = [1]
cuenca = 20
```

```python
# filtrar lead times
df_plot = df[df["lead"].isin(lead_times_evaluar)].copy()
#promediar por lead times
df_plot = (
    df_plot
    .groupby("mes")
    .agg({
        "crpss": "mean",
        "crpss_med": "mean",
        "p5": "mean",
        "p25": "mean",
        "p75": "mean",
        "p95": "mean"
    })
    .reset_index()
    .sort_values("mes")
)
```

```python
stats = []

for _, row in df_plot.iterrows():
    stats.append({
        "label": int(row["mes"]),
        "med": row["crpss_med"],
        "q1": row["p25"],
        "q3": row["p75"],
        "whislo": row["p5"],
        "whishi": row["p95"],
        "fliers": []
    })
    
# === gráfico ===
fig, ax = plt.subplots(figsize=(10, 5))

ax.bxp(stats, showfliers=False)

# línea CRPSS puntual
ax.plot(
    df_plot["mes"],
    df_plot["crpss"],
    marker="o"
)

ax.set_xlabel("Mes")
ax.set_ylabel("CRPSS")
ax.set_title(f"CRPSS mensual - Cuenca {cuenca} - Leads {lead_times_evaluar}")
ax.grid(True)

plt.tight_layout()

# guardar
plt.savefig(ruta_salida / f"crpss_mensual_cuenca_{cuenca}.png", dpi=300)

plt.show()
```
