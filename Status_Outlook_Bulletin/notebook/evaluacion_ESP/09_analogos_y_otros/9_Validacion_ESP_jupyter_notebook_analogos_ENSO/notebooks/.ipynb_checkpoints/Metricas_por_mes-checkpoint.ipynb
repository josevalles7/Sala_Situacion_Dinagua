```python
#importacion

import pandas as pd
import numpy as np
from scipy.stats import pearsonr
from pathlib import Path
from tqdm import tqdm
```

```python
# Rutas

ruta_carpeta = Path(
    "./Resultados/Tablas_de_resultados"
)

ruta_salida = Path(
    "./Resultados/Tabla_de_metrics_por_mes"
)
ruta_salida.mkdir(parents=True, exist_ok=True)
```

Bloque de definición de métricas deterministicas que comparan el promedio de los ensambles con el simulado con clima histórico:

```python
def nash_sutcliffe(q_obs, q_sim):
    return 1 - np.sum((q_obs - q_sim)**2) / np.sum((q_obs - np.mean(q_obs))**2)

def pbias(q_obs, q_sim):
    return 100 * np.sum(q_sim - q_obs) / np.sum(q_obs)

def cv_rmse(q_obs, q_sim):
    rmse = np.sqrt(np.mean((q_sim - q_obs)**2))
    return rmse / np.mean(q_obs)

def pearson_corr(q_obs, q_sim):
    return pearsonr(q_obs, q_sim)[0]
```

Definición del calculo del crpss_ad (métrica probabilistica)

```python
def crps_ensemble(ens, obs):
    # primero ordenamos los ensambles de menor a mayor
    ens = np.sort(ens)
    #calcula el promedio de la distancia entre cada ensamble y el observado
    term1 = np.mean(np.abs(ens - obs))
    #acá vemos a crear un lista de "columnas" y una lista de "filas" y vamos crear una matriz con la diferencia entre cada columna y fila
    #obtenemos la media de esta matriz y así medimos la dispersión entre los ensambles
    term2 = 0.5 * np.mean(np.abs(ens[:, None] - ens[None, :]))
    #este es el valor de crps
    return term1 - term2


def crpss_ad(df_lead, ens_cols):
    
    #df_lead se va a crear más adelante al correr las funciones de las métricas. Es un df que organiza los resultados de las tables por leadtime
    #clonar
    df_lead = df_lead.copy()
    #pasar fechas a datetime
    df_lead["fecha_pronostico"] = pd.to_datetime(df_lead["fecha_pronostico"])

    # Crear columna de mes
    df_lead["mes"] = df_lead["fecha_pronostico"].dt.month

    # --- Climatología como ensamble (por mes) ---
    #para poder crear el crps de referencia con los valores de la climatologia, se organiza la climatologia
    clim_dict = (
        df_lead.groupby("mes")["QC_hist"]
        .apply(lambda x: x.dropna().values)
        .to_dict()
    ) #diccionario en donde la llave es el mes y q_hist es el valor

    # Listas vacías para ir guardando los errores calculados en cada fila.
    crps_model = []
    crps_ref = []

    #Empieza a recorrer el DataFrame fila por fila.
    for _, row in df_lead.iterrows():
        #Toma qc_hist de esa fila.
        obs = row["QC_hist"]
        
        #Si no hay dato, salta esta fila.
        if np.isnan(obs):
            continue

        # --- Ensamble del modelo ---
        #Toma solo las columnas que empiezan con "ens_", quita nulos y lo vuelve un array.
        ens = row[ens_cols].dropna().values
        #Si el ensamble tiene menos de dos membros, además de algo estar mal no se puede calcular dispersión, entonces se saltea
        if len(ens) < 2:
            continue

        # --- Ensamble climatológico (referencia para avaliar) ---
        mes = row["mes"]
        #Busca en el diccionario de antes todos los eneros (o el mes que toque) que existen.
        clim_ens = clim_dict.get(mes, None)

        # Si no hay datos históricos suficientes para ese mes, salta.
        if clim_ens is None or len(clim_ens) < 2:
            continue

        # quita el valor observado actual de la lista histórica
        clim_ens = clim_ens[clim_ens != obs]
        if len(clim_ens) < 2:
            continue

        # --- aplicar el CRPS a los ensambles del modelo y de referencia ---
        crps_model.append(crps_ensemble(ens, obs))
        crps_ref.append(crps_ensemble(clim_ens, obs))

    return 1 - np.mean(crps_model) / np.mean(crps_ref)
```

Definición del cálculo del brier escore para diferentes percentiles

```python
def brier_score_q10(df_lead, ens_cols):
    # toma el df organizado por leadtime.
    df_lead = df_lead.copy()
    # Creamos una columna temporal con el número del mes (1 al 12).
    df_lead["mes"] = pd.to_datetime(df_lead["fecha_pronostico"]).dt.month

    # Agrupa por mes, toma el histórico, quita nulos y calcula el percentil X de ESE mes.
    q10_dict = df_lead.groupby("mes")["QC_hist"].apply(lambda x: np.percentile(x.dropna(), 10)).to_dict()

    #contenedor para resultados
    bs = []

    for _, row in df_lead.iterrows():
        #para cada fila obtener los valores de los ensambles
        ens = row[ens_cols].dropna().values
        #para cada fila obtener el valor simluado histórico
        obs = row["QC_hist"]

        # identificamos qué mes estamos evaluando en esta fila específica.
        mes = row["mes"]

        # filtro por si algun mes no tiene ensambles suficientes o simulado histórico
        if len(ens) < 2 or np.isnan(obs):
            continue

        # acá vamos obtener el valor de referencia Q"x" calculado para el mes específico que se está evaluando
        q10 = q10_dict[mes]

        # empezamos a calcular cuantos ensambles caen en el umbral
        p = np.mean(ens < q10)

        #el observado, cae dentro del umbral?
        o = int(obs < q10)

        # guardamos el error cuadrático del pronóstico para este mes específico.
        bs.append((p - o) ** 2)
                
    # el brier score final es el promedio de todos esos errores cuadráticos mensuales.
    return np.mean(bs)

def brier_score_q10_q25(df_lead, ens_cols):

    df_lead = df_lead.copy()
    df_lead["mes"] = pd.to_datetime(df_lead["fecha_pronostico"]).dt.month

    q10_dict = df_lead.groupby("mes")["QC_hist"].apply(lambda x: np.percentile(x.dropna(), 10)).to_dict()
    q25_dict = df_lead.groupby("mes")["QC_hist"].apply(lambda x: np.percentile(x.dropna(), 25)).to_dict()

    bs = []

    for _, row in df_lead.iterrows():
        ens = row[ens_cols].dropna().values
        obs = row["QC_hist"]

        mes = row["mes"]

        if len(ens) < 2 or np.isnan(obs):
            continue

        q10 = q10_dict[mes]
        q25 = q25_dict[mes]

        p = np.mean((ens >= q10) & (ens < q25))

        o = int((obs >= q10) & (obs < q25))

        bs.append((p - o) ** 2)

    return np.mean(bs)


def brier_score_q25_q75(df_lead, ens_cols):

    df_lead = df_lead.copy()
    df_lead["mes"] = pd.to_datetime(df_lead["fecha_pronostico"]).dt.month

    q25_dict = df_lead.groupby("mes")["QC_hist"].apply(lambda x: np.percentile(x.dropna(), 25)).to_dict()
    q75_dict = df_lead.groupby("mes")["QC_hist"].apply(lambda x: np.percentile(x.dropna(), 75)).to_dict()

    bs = []

    for _, row in df_lead.iterrows():
        ens = row[ens_cols].dropna().values
        obs = row["QC_hist"]

        mes = row["mes"]

        if len(ens) < 2 or np.isnan(obs):
            continue

        q25 = q25_dict[mes]
        q75 = q75_dict[mes]

        p = np.mean((ens >= q25) & (ens <= q75))

        o = int((obs >= q25) & (obs <= q75))

        bs.append((p - o) ** 2)

    return np.mean(bs)

    
def brier_score_q75_q90(df_lead, ens_cols):

    df_lead = df_lead.copy()
    df_lead["mes"] = pd.to_datetime(df_lead["fecha_pronostico"]).dt.month

    q75_dict = df_lead.groupby("mes")["QC_hist"].apply(lambda x: np.percentile(x.dropna(), 75)).to_dict()
    q90_dict = df_lead.groupby("mes")["QC_hist"].apply(lambda x: np.percentile(x.dropna(), 90)).to_dict()

    bs = []

    for _, row in df_lead.iterrows():
        ens = row[ens_cols].dropna().values
        obs = row["QC_hist"]

        mes = row["mes"]

        if len(ens) < 2 or np.isnan(obs):
            continue

        q75 = q75_dict[mes]
        q90 = q90_dict[mes]

        p = np.mean((ens > q75) & (ens <= q90))

        o = int((obs > q75) & (obs <= q90))

        bs.append((p - o) ** 2)

    return np.mean(bs)



def brier_score_q90(df_lead, ens_cols):

    df_lead = df_lead.copy()
    df_lead["mes"] = pd.to_datetime(df_lead["fecha_pronostico"]).dt.month

    q90_dict = df_lead.groupby("mes")["QC_hist"].apply(lambda x: np.percentile(x.dropna(), 90)).to_dict()

    bs = []

    for _, row in df_lead.iterrows():
        ens = row[ens_cols].dropna().values
        obs = row["QC_hist"]

        mes = row["mes"]

        if len(ens) < 2 or np.isnan(obs):
            continue

        q90 = q90_dict[mes]

        p = np.mean(ens >= q90)
                    
        o = int(obs > q90)

        bs.append((p - o) ** 2)

    return np.mean(bs)
```

Loop sobre las tablas de resultados para calcular las métricas:

```python
archivos = list(ruta_carpeta.glob("*.csv"))
for ruta_csv in tqdm(archivos, desc="Procesando cuencas"):
    
    df = pd.read_csv(ruta_csv)

    df["fecha_emision"] = pd.to_datetime(df["fecha_emision"])
    df["fecha_pronostico"] = pd.to_datetime(df["fecha_pronostico"])

    df["mes"] = df["fecha_pronostico"].dt.month

    ens_cols = [c for c in df.columns if c.startswith("ens_") and c != "ens_prom"]

    resultados = []

    for (lead, mes), df_sub in df.groupby(["lead", "mes"]):

        q_obs = df_sub["QC_hist"].values
        q_sim = df_sub["ens_prom"].values

        metricas = {
            "lead": lead,
            "NS": nash_sutcliffe(q_obs, q_sim),
            "Pbias": pbias(q_obs, q_sim),
            "CV": cv_rmse(q_obs, q_sim),
            "Pearson": pearson_corr(q_obs, q_sim),
            "CRPSSad": crpss_ad(df_sub, ens_cols),
            "Brier_Q10": brier_score_q10(df_sub, ens_cols),
            "Brier_Q10_Q25": brier_score_q10_q25(df_sub, ens_cols),
            "Brier_Q25_Q75": brier_score_q25_q75(df_sub, ens_cols),
            "Brier_Q75_Q90": brier_score_q75_q90(df_sub, ens_cols),
            "Brier_Q90": brier_score_q90(df_sub, ens_cols),
            "N_det": np.sum(~np.isnan(q_sim)),
            "N_prob": np.sum(df_sub[ens_cols].notna().sum(axis=1) >= 2)
        }
        
        resultados.append(metricas)

    df_out = pd.DataFrame(resultados)

    df_out.to_csv(
        ruta_salida / f"{ruta_csv.stem}_metricas_por_mes.csv",
        index=False
    )
```
