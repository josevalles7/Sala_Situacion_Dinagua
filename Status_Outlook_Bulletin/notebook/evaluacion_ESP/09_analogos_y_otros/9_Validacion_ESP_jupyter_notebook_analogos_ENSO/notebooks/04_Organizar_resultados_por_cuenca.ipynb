```python
import pandas as pd
import os
```

```python
ruta = r"../Resultados/Tablas_de_metricas_por_lead"

dfs = []
```

```python
for archivo in os.listdir(ruta):
    if archivo.endswith(".csv"):
        path = os.path.join(ruta, archivo)
        
        df = pd.read_csv(path)
        df = df.set_index("lead")

        df = df.drop(columns=["NS","Pbias","CV","Pearson","CRPSSad","Brier_Q10","Brier_Q10_Q25","Brier_Q25_Q75","Brier_Q75_Q90","Brier_Q90","N_det","N_prob"])
        
        cuenca = os.path.basename(archivo).split("_")[0]
        
        df = df.rename(columns={"CRPSSad": cuenca})
        
        dfs.append(df)

# Concatenar por columnas (alineando por lead)
print (dfs)
df_final = pd.concat(dfs, axis=1)

df_final = df_final.transpose()

# Guardar
df_final.to_csv(os.path.join(r"../Resultados/Metricas_por_cuenca/datos_CRPSSad_por_cuenca.csv"))
```
